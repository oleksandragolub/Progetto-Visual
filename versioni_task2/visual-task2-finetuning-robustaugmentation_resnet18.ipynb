```python
# =============================================================================
# TASK 2 â€” ROBUST CLASSIFICATION v2 (VERSIONE OTTIMIZZATA)
# =============================================================================

import subprocess
subprocess.run(["pip", "install", "-q", "albumentations"], check=True)

# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# IMPORTS
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
import random, os
import numpy as np
import matplotlib.pyplot as plt
from tqdm.auto import tqdm
from PIL import Image
from dataclasses import dataclass
from sklearn.metrics import accuracy_score

import cv2
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.amp import GradScaler, autocast
from torchvision import datasets, models
from torch.utils.data import DataLoader, Dataset

import albumentations as A
from albumentations.pytorch import ToTensorV2

print(f"âœ“ PyTorch  : {torch.__version__}")
print(f"âœ“ OpenCV   : {cv2.__version__}")
print(f"âœ“ CUDA     : {torch.cuda.is_available()}")


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# CONFIGURAZIONE
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
@dataclass
class Config:
    # â”€â”€ Paths â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    data_dir       = "/kaggle/input/datasets/andreaspagnolo/visual-exam-dataset-2/visual_dataset"
    test_deg_dir   = "/kaggle/input/datasets/andreaspagnolo/visual-exam-dataset-3/test_denoised"
    test_clean_dir = "/kaggle/input/datasets/andreaspagnolo/visual-exam-dataset/visual_dataset/test"
    model_path     = "/kaggle/input/models/andreaspagnolo2acc/resnet18-model-recalcolate/pytorch/default/1/best_model_resnet18_recalcolate.pth"
    save_path      = "best_model_task2_v2.pth"

    # â”€â”€ Training â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    batch_size     = 48
    epochs         = 45
    lr             = 5e-5
    weight_decay   = 5e-4

    # â”€â”€ Augmentation â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    noise_prob     = 0.75
    progressive    = True

    # â”€â”€ Fine-tuning â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    freeze_epochs  = 3     # freeze backbone, train only fc

    # â”€â”€ Regularization â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    label_smoothing = 0.1
    mixup_alpha    = 0.2
    mixup_prob     = 0.3

    # â”€â”€ Val set rumoroso â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    noisy_val_copies    = 1
    noisy_val_intensity = 0.75

    # â”€â”€ Early stopping â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    patience       = 10

cfg = Config()

def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
set_seed(42)
print(f"âœ“ Device: {device}")


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# MODELLO â€” plain ResNet18 (stesso di Task 1)
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def load_pretrained(path: str, num_classes: int, device: torch.device) -> nn.Module:
    """
    Builds the exact same architecture used in Task 1
    (torchvision ResNet18, no pretrained weights, custom fc)
    and loads the saved state_dict.
    """
    model = models.resnet18(weights=None)
    model.fc = nn.Linear(model.fc.in_features, num_classes)

    state = torch.load(path, map_location=device, weights_only=False)
    # Task 1 saved bare state_dict (torch.save(model.state_dict(), ...))
    # but handle the case where it was wrapped in a dict just in case
    if isinstance(state, dict) and "model_state_dict" in state:
        state = state["model_state_dict"]

    model.load_state_dict(state, strict=True)
    model = model.to(device)
    print(f"âœ“ ResNet18 caricato da '{path}'")
    print(f"  Parametri totali : {sum(p.numel() for p in model.parameters()):,}")
    return model


def set_trainable(model: nn.Module, freeze_backbone: bool):
    """Freeze everything except the fc layer, or unfreeze all."""
    for name, p in model.named_parameters():
        p.requires_grad = True if "fc" in name else (not freeze_backbone)
    n_train = sum(p.numel() for p in model.parameters() if p.requires_grad)
    n_tot   = sum(p.numel() for p in model.parameters())
    print(f"Backbone {'FROZEN' if freeze_backbone else 'UNFROZEN'} | "
          f"Trainable: {n_train:,} / {n_tot:,}")


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# AUGMENTATION PIPELINE
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
MEAN = [0.485, 0.456, 0.406]
STD  = [0.229, 0.224, 0.225]


def build_noisy_transform(intensity_factor: float = 1.0) -> A.Compose:
    i = max(0.2, intensity_factor)

    return A.Compose([
        A.Resize(256, 256),
        A.RandomResizedCrop(size=(224, 224), scale=(0.4, 1.0)),
        A.HorizontalFlip(p=0.5),
        A.Rotate(limit=20, p=0.5),

        A.OneOf([
            A.GaussNoise(std_range=(0.02 * i, 0.12 * i), p=1),
            A.ISONoise(color_shift=(0.01, 0.05 * i),
                       intensity=(0.1 * i, 0.8 * i), p=1),
            A.MultiplicativeNoise(
                multiplier=(1.0 - 0.18 * i, 1.0 + 0.18 * i), p=1),
            A.ShotNoise(scale_range=(0.05 * i, 0.40 * i), p=1),
        ], p=0.80),

        A.OneOf([
            A.GaussianBlur(blur_limit=(3, max(3, int(11 * i))), p=1),
            A.MotionBlur(blur_limit=(3, max(3, int(9 * i))), p=1),
            A.Defocus(radius=(1, max(1, int(5 * i))),
                      alias_blur=(0.1, 0.5), p=1),
            A.MedianBlur(blur_limit=(3, 7), p=1),
            A.ZoomBlur(max_factor=1.0 + 0.15 * i, p=1),
        ], p=0.70),

        A.OneOf([
            A.ImageCompression(
                quality_range=(max(5, int(55 - 30 * i)), 90), p=1),
            A.Downscale(scale_range=(0.35, 0.85), p=1),
            A.Posterize(num_bits=(4, 7), p=1),
            A.Solarize(p=1),
        ], p=0.50),

        A.OneOf([
            A.RandomBrightnessContrast(
                brightness_limit=0.45 * i,
                contrast_limit=0.45 * i, p=1),
            A.HueSaturationValue(
                hue_shift_limit=int(30 * i),
                sat_shift_limit=int(45 * i),
                val_shift_limit=int(30 * i), p=1),
            A.CLAHE(clip_limit=4.0 * i, tile_grid_size=(8, 8), p=1),
            A.RandomGamma(gamma_limit=(60, 140), p=1),
            A.ToGray(p=1),
        ], p=0.50),

        A.OneOf([
            A.OpticalDistortion(distort_limit=0.35 * i, p=1),
            A.GridDistortion(num_steps=5, distort_limit=0.3 * i, p=1),
            A.ElasticTransform(alpha=1, sigma=10, p=1),
            A.Perspective(scale=(0.05, 0.12 * i), p=1),
        ], p=0.35),

        A.OneOf([
            A.RandomFog(fog_coef_range=(0.05 * i, 0.35 * i),
                        alpha_coef=0.1, p=1),
            A.RandomShadow(num_shadows_limit=(1, 2),
                           shadow_dimension=5, p=1),
            A.RandomRain(slant_range=(-10, 10), drop_length=10,
                         drop_width=1, blur_value=3, p=1),
        ], p=0.20),

        A.Normalize(mean=MEAN, std=STD),
        ToTensorV2(),
    ])


def build_clean_transform() -> A.Compose:
    return A.Compose([
        A.Resize(256, 256),
        A.RandomResizedCrop(size=(224, 224), scale=(0.4, 1.0)),
        A.HorizontalFlip(p=0.5),
        A.Rotate(limit=20, p=0.5),
        A.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.2, p=0.5),
        A.Normalize(mean=MEAN, std=STD),
        ToTensorV2(),
    ])


def build_val_transform() -> A.Compose:
    return A.Compose([
        A.Resize(256, 256),
        A.CenterCrop(224, 224),
        A.Normalize(mean=MEAN, std=STD),
        ToTensorV2(),
    ])


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# DATASET CLASSES
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
class MixedDataset(Dataset):
    def __init__(self, base_ds, tf_clean, tf_noisy, noise_prob=0.75):
        self.base_ds    = base_ds
        self.tf_clean   = tf_clean
        self.tf_noisy   = tf_noisy
        self.noise_prob = noise_prob

    def __len__(self):
        return len(self.base_ds)

    def __getitem__(self, idx):
        path, label = self.base_ds.samples[idx]
        img = np.array(Image.open(path).convert("RGB"))
        if random.random() < self.noise_prob:
            return self.tf_noisy(image=img)["image"], label
        else:
            return self.tf_clean(image=img)["image"], label


class NoisyValDataset(Dataset):
    def __init__(self, base_ds, tf_val, tf_noisy_moderate, n_noisy_copies=1):
        self.entries = []
        for path, label in tqdm(base_ds.samples,
                                desc="Building NoisyValDataset", leave=False):
            img = np.array(Image.open(path).convert("RGB"))
            self.entries.append((img, label, tf_val))
            for _ in range(n_noisy_copies):
                self.entries.append((img, label, tf_noisy_moderate))

    def __len__(self):
        return len(self.entries)

    def __getitem__(self, idx):
        img, label, tf = self.entries[idx]
        return tf(image=img.copy())["image"], label


class TransformDataset(Dataset):
    def __init__(self, base_ds, transform):
        self.base_ds   = base_ds
        self.transform = transform

    def __len__(self):
        return len(self.base_ds)

    def __getitem__(self, idx):
        path, label = self.base_ds.samples[idx]
        img = np.array(Image.open(path).convert("RGB"))
        return self.transform(image=img)["image"], label


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# MIXUP
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def mixup_data(x, y, alpha=0.2):
    lam = np.random.beta(alpha, alpha) if alpha > 0 else 1.0
    idx = torch.randperm(x.size(0), device=x.device)
    return lam * x + (1 - lam) * x[idx], y, y[idx], lam

def mixup_loss(criterion, pred, ya, yb, lam):
    return lam * criterion(pred, ya) + (1 - lam) * criterion(pred, yb)


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# TRAINING LOOP
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def train_one_epoch(model, loader, criterion, optimizer, scaler, device, cfg):
    model.train()
    loss_sum = 0.0
    preds_all, labels_all = [], []

    for inputs, labels in tqdm(loader, desc="Train", leave=False):
        inputs, labels = inputs.to(device), labels.to(device)
        use_mixup = random.random() < cfg.mixup_prob
        if use_mixup:
            inputs, ya, yb, lam = mixup_data(inputs, labels, cfg.mixup_alpha)

        optimizer.zero_grad()
        with autocast(device_type=device.type):
            out  = model(inputs)
            loss = mixup_loss(criterion, out, ya, yb, lam) if use_mixup \
                   else criterion(out, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        loss_sum += loss.item() * inputs.size(0)
        if not use_mixup:
            preds_all.extend(out.argmax(1).cpu().numpy())
            labels_all.extend(labels.cpu().numpy())

    acc = accuracy_score(labels_all, preds_all) if labels_all else 0.0
    return loss_sum / len(loader.dataset), acc


@torch.no_grad()
def evaluate(model, loader, criterion, device):
    model.eval()
    loss_sum = 0.0
    preds, labels = [], []
    for inputs, labs in tqdm(loader, desc="Val  ", leave=False):
        inputs, labs = inputs.to(device), labs.to(device)
        with autocast(device_type=device.type):
            out  = model(inputs)
            loss = criterion(out, labs)
        loss_sum += loss.item() * inputs.size(0)
        preds.extend(out.argmax(1).cpu().numpy())
        labels.extend(labs.cpu().numpy())
    return loss_sum / len(loader.dataset), accuracy_score(labels, preds)


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# MAIN TRAINING
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def make_optimizer_and_scheduler(model, cfg, frozen: bool):
    params    = filter(lambda p: p.requires_grad, model.parameters()) \
                if frozen else model.parameters()
    optimizer = optim.AdamW(params, lr=cfg.lr, weight_decay=cfg.weight_decay)
    remaining = cfg.freeze_epochs if frozen else cfg.epochs - cfg.freeze_epochs
    scheduler = optim.lr_scheduler.CosineAnnealingLR(
        optimizer, T_max=max(1, remaining), eta_min=1e-7)
    return optimizer, scheduler


def train(cfg):
    print("=" * 80)
    print("TASK 2 v2 â€” ROBUST FINE-TUNING (plain ResNet18 da Task 1)")
    print("=" * 80)

    train_base = datasets.ImageFolder(os.path.join(cfg.data_dir, "train"))
    val_base   = datasets.ImageFolder(os.path.join(cfg.data_dir, "valid"))
    n_classes  = len(train_base.classes)
    print(f"Classi: {n_classes} | Train: {len(train_base)} | Val: {len(val_base)}")

    model     = load_pretrained(cfg.model_path, n_classes, device)
    criterion = nn.CrossEntropyLoss(label_smoothing=cfg.label_smoothing)
    scaler    = GradScaler(device.type)

    set_trainable(model, freeze_backbone=True)
    optimizer, scheduler = make_optimizer_and_scheduler(model, cfg, frozen=True)

    tf_val            = build_val_transform()
    tf_noisy_moderate = build_noisy_transform(intensity_factor=cfg.noisy_val_intensity)

    print(f"\nPreparazione NoisyValDataset "
          f"(n_copies={cfg.noisy_val_copies}, intensity={cfg.noisy_val_intensity})...")
    noisy_val_ds = NoisyValDataset(val_base, tf_val, tf_noisy_moderate,
                                   n_noisy_copies=cfg.noisy_val_copies)

    dl_kwargs        = dict(batch_size=cfg.batch_size, num_workers=4,
                            pin_memory=True, persistent_workers=True)
    noisy_val_loader = DataLoader(noisy_val_ds, shuffle=False, **dl_kwargs)
    clean_val_loader = DataLoader(
        TransformDataset(val_base, tf_val), shuffle=False, **dl_kwargs)

    history = {k: [] for k in ["train_loss", "train_acc",
                                "noisy_val_loss", "noisy_val_acc",
                                "clean_val_acc", "lr"]}
    best_noisy_acc = 0.0
    patience_cnt   = 0

    for epoch in range(cfg.epochs):
        print(f"\n{'â”€'*80}")
        current_lr = optimizer.param_groups[0]["lr"]
        print(f"Epoch {epoch+1:02d}/{cfg.epochs}  (lr={current_lr:.2e})")

        if epoch == cfg.freeze_epochs:
            print("ðŸ”“  SBLOCCO BACKBONE")
            set_trainable(model, freeze_backbone=False)
            optimizer, scheduler = make_optimizer_and_scheduler(
                model, cfg, frozen=False)

        if cfg.progressive:
            elapsed = max(0, epoch - cfg.freeze_epochs)
            total   = cfg.epochs - cfg.freeze_epochs
            t       = min(elapsed / max(1, total * 0.7), 1.0)
            intensity = 0.25 + 0.75 * t
        else:
            intensity = 1.0

        tf_noisy = build_noisy_transform(intensity)
        tf_clean = build_clean_transform()

        train_ds     = MixedDataset(train_base, tf_clean, tf_noisy,
                                    noise_prob=cfg.noise_prob)
        train_loader = DataLoader(train_ds, shuffle=True, **dl_kwargs)

        t_loss, t_acc   = train_one_epoch(
            model, train_loader, criterion, optimizer, scaler, device, cfg)
        nv_loss, nv_acc = evaluate(model, noisy_val_loader, criterion, device)
        _,       cv_acc = evaluate(model, clean_val_loader, criterion, device)

        scheduler.step()

        history["train_loss"].append(t_loss)
        history["train_acc"].append(t_acc)
        history["noisy_val_loss"].append(nv_loss)
        history["noisy_val_acc"].append(nv_acc)
        history["clean_val_acc"].append(cv_acc)
        history["lr"].append(current_lr)

        print(f"  Train       â†’ loss={t_loss:.4f}  acc={t_acc:.4f}")
        print(f"  Val (noisy) â†’ loss={nv_loss:.4f}  acc={nv_acc:.4f}"
              f"  â† MODEL SELECTION  [intensity={intensity:.2f}]")
        print(f"  Val (clean) â†’ acc={cv_acc:.4f}  (monitoraggio)")

        if nv_acc > best_noisy_acc:
            best_noisy_acc = nv_acc
            torch.save({"epoch": epoch,
                        "model_state_dict": model.state_dict(),
                        "best_noisy_val_acc": best_noisy_acc,
                        "intensity_at_best": intensity}, cfg.save_path)
            print(f"  âœ“ Best model  (noisy val acc = {best_noisy_acc:.4f})")
            patience_cnt = 0
        else:
            patience_cnt += 1
            print(f"  âš  No improvement [{patience_cnt}/{cfg.patience}]")

        if patience_cnt >= cfg.patience:
            print(f"\nâš  Early stopping all'epoca {epoch+1}")
            break

    print(f"\nâœ“ Training terminato! Best noisy val acc: {best_noisy_acc:.4f}")
    model.load_state_dict(
        torch.load(cfg.save_path, weights_only=False)["model_state_dict"])
    return model, history


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# TTA â€” 10 varianti noise-aware
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def build_tta_transforms() -> list:
    norm = [A.Normalize(mean=MEAN, std=STD), ToTensorV2()]
    return [
        A.Compose([A.Resize(256, 256), A.CenterCrop(224, 224), *norm]),
        A.Compose([A.Resize(256, 256), A.CenterCrop(224, 224),
                   A.HorizontalFlip(p=1), *norm]),
        A.Compose([A.Resize(256, 256), A.CenterCrop(224, 224),
                   A.Sharpen(alpha=(0.3, 0.5), lightness=(0.8, 1.2), p=1), *norm]),
        A.Compose([A.Resize(256, 256), A.CenterCrop(224, 224),
                   A.HorizontalFlip(p=1),
                   A.Sharpen(alpha=(0.5, 0.7), lightness=(0.9, 1.1), p=1), *norm]),
        A.Compose([A.Resize(256, 256), A.CenterCrop(224, 224),
                   A.CLAHE(clip_limit=3.0, tile_grid_size=(8, 8), p=1), *norm]),
        A.Compose([A.Resize(256, 256), A.CenterCrop(224, 224),
                   A.HorizontalFlip(p=1),
                   A.CLAHE(clip_limit=3.0, tile_grid_size=(8, 8), p=1), *norm]),
        A.Compose([A.Resize(288, 288), A.CenterCrop(224, 224), *norm]),
        A.Compose([A.Resize(288, 288), A.CenterCrop(224, 224),
                   A.HorizontalFlip(p=1), *norm]),
        A.Compose([A.Resize(256, 256), A.RandomCrop(224, 224), *norm]),
        A.Compose([A.Resize(256, 256), A.CenterCrop(224, 224),
                   A.UnsharpMask(blur_limit=(3, 5), sigma_limit=(0.0, 1.0),
                                 alpha=(0.2, 0.5), p=1), *norm]),
    ]


@torch.no_grad()
def predict_tta(model, img_np: np.ndarray, tta_tfs: list, device) -> int:
    model.eval()
    probs_list = []
    for tf in tta_tfs:
        t = tf(image=img_np.copy())["image"].unsqueeze(0).to(device)
        with autocast(device_type=device.type):
            p = F.softmax(model(t), dim=1)
        probs_list.append(p.cpu().numpy())
    return int(np.argmax(np.mean(probs_list, axis=0)[0]))


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# VALUTAZIONE TEST SET
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def evaluate_test(model, test_dir, device, use_tta=True, label=""):
    tag = f"[{label}] " if label else ""
    print(f"\n{'='*80}")
    print(f"{tag}TTA={'ON' if use_tta else 'OFF'}")
    print(f"Dir: {test_dir}")
    print("=" * 80)

    test_ds = datasets.ImageFolder(test_dir)
    tf_val  = build_val_transform()
    tta_tfs = build_tta_transforms() if use_tta else None

    preds  = []
    labels = [l for _, l in test_ds.samples]

    for path, _ in tqdm(test_ds.samples, desc="Test"):
        img = np.array(Image.open(path).convert("RGB"))
        if use_tta:
            pred = predict_tta(model, img, tta_tfs, device)
        else:
            t = tf_val(image=img)["image"].unsqueeze(0).to(device)
            with torch.no_grad(), autocast(device_type=device.type):
                pred = int(model(t).argmax(1).cpu())
        preds.append(pred)

    acc = accuracy_score(labels, preds) * 100
    print(f"Accuracy  : {acc:.2f}%")
    print(f"Baseline  : 74.60%")
    print(f"Î” Baseline: {acc - 74.60:+.2f}%")
    return acc


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# PLOT
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
def plot_history(history):
    ep = range(1, len(history["train_loss"]) + 1)
    fig, axes = plt.subplots(1, 3, figsize=(20, 5))

    axes[0].plot(ep, history["train_loss"],     "o-", label="Train loss")
    axes[0].plot(ep, history["noisy_val_loss"], "s-", label="Noisy val loss")
    axes[0].set_title("Loss"); axes[0].legend(); axes[0].grid(True)
    axes[0].set_xlabel("Epoch")

    axes[1].plot(ep, history["train_acc"],     "o-", label="Train acc")
    axes[1].plot(ep, history["noisy_val_acc"], "s-",
                 label="Noisy val acc (model sel)", linewidth=2)
    axes[1].plot(ep, history["clean_val_acc"], "^--",
                 label="Clean val acc (monitor)", alpha=0.7)
    axes[1].axhline(y=0.746, color="red", linestyle=":", alpha=0.5,
                    label="Baseline (74.6%)")
    axes[1].set_title("Accuracy"); axes[1].legend(); axes[1].grid(True)
    axes[1].set_xlabel("Epoch")

    ax2b = axes[2].twinx()
    axes[2].plot(ep, history["noisy_val_acc"], "s-",
                 color="tab:blue", label="Noisy val acc")
    ax2b.plot(ep, history["lr"], ".-",
              color="tab:orange", alpha=0.5, label="LR")
    axes[2].set_title("LR monotono vs Noisy Val Acc")
    axes[2].set_ylabel("Accuracy", color="tab:blue")
    ax2b.set_ylabel("Learning Rate", color="tab:orange")
    axes[2].grid(True); axes[2].set_xlabel("Epoch")

    for ax in axes:
        ax.axvline(x=cfg.freeze_epochs + 0.5, color="green",
                   linestyle="--", alpha=0.4, label="Unfreeze")

    plt.tight_layout()
    plt.savefig("training_history_task2_v2.png", dpi=150)
    plt.show()
    print("âœ“ Plot salvato: training_history_task2_v2.png")


# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
# ENTRY POINT
# â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
if __name__ == "__main__" or True:

    # â”€â”€ TRAINING â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    model, history = train(cfg)
    plot_history(history)

    # â”€â”€ VALUTAZIONE FINALE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n" + "=" * 80)
    print("VALUTAZIONE FINALE")
    print("=" * 80)

    deg_results = {}
    for label, use_tta in [
        ("Degradato | no TTA", False),
        ("Degradato | TTA",    True),
    ]:
        deg_results[label] = evaluate_test(
            model, cfg.test_deg_dir, device,
            use_tta=use_tta, label=label)

    # Clean: no TTA, solo reference
    clean_acc = evaluate_test(
        model, cfg.test_clean_dir, device,
        use_tta=False, label="Pulito | no TTA (reference)")

    # â”€â”€ RIEPILOGO FINALE â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€
    print("\n" + "=" * 80)
    print("RIEPILOGO FINALE v2")
    print("=" * 80)
    print(f"{'Setup':<42} | {'Accuracy':>9} | {'Î” Baseline':>11} | {'Î” vs v1 (79%)':>13}")
    print("-" * 82)

    v1_best      = 79.00
    best_deg_acc = max(deg_results.values())   # best calcolato SOLO sul degradato

    for label, acc in deg_results.items():
        star = " â˜…" if acc == best_deg_acc else "  "
        print(f"{label:<42} | {acc:>8.2f}% | {acc - 74.60:>+10.2f}% | "
              f"{acc - v1_best:>+12.2f}%{star}")

    print("-" * 82)
    print(f"{'Pulito | no TTA (reference)':<42} | {clean_acc:>8.2f}% | "
          f"{clean_acc - 74.60:>+10.2f}% | {clean_acc - v1_best:>+12.2f}%  (ref)")

    print("=" * 80)
    print(f"\nBest acc su degradato : {best_deg_acc:.2f}%")
    print(f"Î” vs baseline (74.60%): {best_deg_acc - 74.60:+.2f}%")
    print(f"Î” vs v1 (79.00%)      : {best_deg_acc - v1_best:+.2f}%")
    print(f"\nClean (reference only): {clean_acc:.2f}%")
```
