```python
BASE = "/content/drive/MyDrive/progetto_visual/codice/extracted"     # percorso alla cartella principale che contiene il contenuto dei zip
TRAIN_DIR = os.path.join(BASE, "train", "train")                     # cartella che contiene le immagini di training
VAL_DIR   = os.path.join(BASE, "valid", "valid")                     # cartella che contiene le immagini di validation
TEST_DIR  = os.path.join(BASE, "test", "test")                       # cartella che contiene le immagini di test
TESTD_DIR = os.path.join(BASE, "test_degradato", "test_degradato")   # cartella che contiene le immagini di test degradato

LABELS_CSV = "/content/drive/MyDrive/progetto_visual/codice/sports_labels.csv"  # percorso al file delle etichette

# verifica dell’esistenza delle directory
print("train exists:", os.path.exists(TRAIN_DIR))
print("valid exists:", os.path.exists(VAL_DIR))
print("test exists:", os.path.exists(TEST_DIR))
print("test_degradato exists:", os.path.exists(TESTD_DIR))
```

```python
# codice per verificare l'organizzazione delle classi nelle cartelle

def peek_classes(d):
    items = sorted([
        x for x in os.listdir(d)
        if os.path.isdir(os.path.join(d, x))
    ])
    print(d)
    print("Numero classi:", len(items))
    print("Classi:", items)

peek_classes(TRAIN_DIR)
peek_classes(VAL_DIR)
peek_classes(TEST_DIR)
peek_classes(TESTD_DIR)
```
